```python
import pandas as pd

df = pd.read_csv("glassdoor_jobs.csv")
df.head()
```

```python
len(df)
```

# Salary Parsing

```python
df["hourly"] = df["Salary Estimate"].apply(lambda x: 1 if "per hour" in x.lower() else 0)
df["employer_provided"] = df["Salary Estimate"].apply(lambda x: 1 if "employer provided salary" in x.lower() else 0)
df[df["employer_provided"] == 1].head(5)
```

```python
df = df[df["Salary Estimate"] != "-1"]
len(df)
```

```python
salary = df["Salary Estimate"].apply(lambda x: x.split("(")[0])
salary
```

```python
removeK = salary.apply(lambda x: x.replace("K", "").replace("$", ""))
removeK
```

```python
removePerHour = removeK.apply(lambda x: x.lower().replace("per hour", "").replace("employer provided salary:", ""))
removePerHour
```

```python
df["min_salary"] = removePerHour.apply(lambda x: int(x.split("-")[0]))
df["max_salary"] = removePerHour.apply(lambda x: int(x.split("-")[1]))
df["avg_salary"] = (df.min_salary + df.max_salary) / 2
df["max_salary"].dtype # check data type
df.head(5)
```

# Remove Rating from Company Name

```python
df["company_txt"] = df.apply(lambda x: x["Company Name"] if x["Rating"] < 0 else x["Company Name"][:-3], axis=1) # specify axis=1 because we are doing it on rows
df.head(5)
```

# Split State and Location

```python
df["job_state"] = df["Location"].apply(lambda x: x.split(",")[1])
df["job_state"]
```

```python
df.job_state.value_counts()
```

```python
# headquarter is located in location
df["same_state"] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df["same_state"]
```

# Get Age of Company

```python
df["age"] = df.Founded.apply(lambda x: -1 if x < 1 else 2022 - x)
df[df["age"] >200]
```

# Parse Job Description Keywords

##### Keyword: python, r studio, spark, aws, excel

```python
df["python_yn"] = df["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0)
df["python_yn"].value_counts()
```

```python
df["r_yn"] = df["Job Description"].apply(lambda x: 1 if "r studio" in x.lower() or "r-studio" in x.lower() else 0)
df["r_yn"].value_counts()
df[df["r_yn"] == 1]["Job Description"][275]
```

```python
df["spark_yn"] = df["Job Description"].apply(lambda x: 1 if "spark" in x.lower() else 0)
df["spark_yn"].value_counts()
```

```python
df["aws_yn"] = df["Job Description"].apply(lambda x: 1 if "aws" in x.lower() else 0)
df["aws_yn"].value_counts()
```

```python
df["excel_yn"] = df["Job Description"].apply(lambda x: 1 if "excel" in x.lower() else 0)
df["excel_yn"].value_counts()
```
